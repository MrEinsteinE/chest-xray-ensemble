```python
# Phase 5: Advanced Architecture - EfficientNet-B3
import os, time, json, numpy as np, pandas as pd, matplotlib.pyplot as plt
from PIL import Image
from tqdm.auto import tqdm
import torch, torch.nn as nn
from torch.utils.data import DataLoader
from torch.amp import autocast, GradScaler
from albumentations import Compose, Resize, Normalize, HorizontalFlip, RandomBrightnessContrast
from albumentations.pytorch import ToTensorV2
from sklearn.metrics import roc_auc_score
from efficientnet_pytorch import EfficientNet
import cv2

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
torch.backends.cudnn.benchmark = True
cv2.setNumThreads(0)

print("="*70)
print("PHASE 5: EFFICIENTNET-B3 ADVANCED TRAINING")
print("="*70)
print(f"Device: {device}")
print(f"CUDA Available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB")
```

```python
class AdvancedDataset(torch.utils.data.Dataset):
    def __init__(self, df, img_dir, transform):
        self.df = df.reset_index(drop=True)
        self.img_dir = img_dir
        self.transform = transform
        self.labels = ['Atelectasis','Consolidation','Infiltration','Pneumothorax','Edema',
                       'Emphysema','Fibrosis','Effusion','Pneumonia','Pleural_Thickening',
                       'Cardiomegaly','Nodule','Mass','Hernia']
    
    def __len__(self): return len(self.df)
    
    def __getitem__(self, i):
        try:
            row = self.df.iloc[i]
            img_path = None
            for j in range(1,13):
                p = os.path.join(self.img_dir, f'images_{j:03d}/images', row['Image Index'])
                if os.path.exists(p): img_path = p; break
            
            img = np.array(Image.open(img_path).convert('RGB'))
            img = self.transform(image=img)['image']
            
            y = np.zeros(len(self.labels), dtype=np.float32)
            if row['Finding Labels'] != "No Finding":
                for k, l in enumerate(self.labels):
                    if l in row['Finding Labels']: y[k] = 1
            
            return img, torch.tensor(y)
        except Exception:
            return None

def safe_collate(batch):
    batch = [b for b in batch if b is not None]
    if not batch: return None
    return torch.utils.data.dataloader.default_collate(batch)
```

```python
extract = "./chest_xray_data/"
data = pd.read_csv(os.path.join(extract, "Data_Entry_2017.csv"))

# Use same split as before for fair comparison
train_df, val_df, test_df = np.split(
    data.sample(frac=1, random_state=42),
    [int(.7*len(data)), int(.85*len(data))]
)

# Transforms with stronger augmentation for B3
tf_train = Compose([
    Resize(224, 224),
    HorizontalFlip(p=0.5),
    RandomBrightnessContrast(0.2, 0.2, p=0.5),
    Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225]),
    ToTensorV2()
])

tf_val = Compose([
    Resize(224, 224),
    Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225]),
    ToTensorV2()
])

train_ds = AdvancedDataset(train_df, extract, tf_train)
val_ds = AdvancedDataset(val_df, extract, tf_val)
test_ds = AdvancedDataset(test_df, extract, tf_val)

# Balanced sampler
labels = ['Atelectasis','Consolidation','Infiltration','Pneumothorax','Edema',
          'Emphysema','Fibrosis','Effusion','Pneumonia','Pleural_Thickening',
          'Cardiomegaly','Nodule','Mass','Hernia']

def create_sampler(df, labels):
    class_counts = np.zeros(len(labels))
    for i, label in enumerate(labels):
        class_counts[i] = df['Finding Labels'].str.contains(label, na=False).sum()
    
    class_weights = 1.0 / (class_counts + 1)
    class_weights = class_weights / class_weights.sum() * len(labels)
    
    sample_weights = []
    for idx, row in df.iterrows():
        labels_present = [i for i, l in enumerate(labels) if l in row['Finding Labels']]
        if labels_present:
            sample_weights.append(max(class_weights[labels_present]))
        else:
            sample_weights.append(1.0)
    
    return torch.utils.data.WeightedRandomSampler(sample_weights, len(sample_weights), replacement=True)

sampler = create_sampler(train_df, labels)

# DataLoaders - reduced batch size for B3 (bigger model)
train_loader = DataLoader(train_ds, batch_size=12, sampler=sampler,  # 12 instead of 16
                          num_workers=0, collate_fn=safe_collate, pin_memory=False)
val_loader = DataLoader(val_ds, batch_size=12, shuffle=False,
                        num_workers=0, collate_fn=safe_collate, pin_memory=False)
test_loader = DataLoader(test_ds, batch_size=12, shuffle=False,
                         num_workers=0, collate_fn=safe_collate, pin_memory=False)

print(f"✅ Data loaded:")
print(f"   Train: {len(train_df):,} images ({len(train_loader)} batches)")
print(f"   Val:   {len(val_df):,} images ({len(val_loader)} batches)")
print(f"   Test:  {len(test_df):,} images ({len(test_loader)} batches)")
```

```python
class EfficientNetB3(nn.Module):
    def __init__(self, num_classes=14, dropout=0.4):
        super().__init__()
        # B3: 12M parameters (vs B0: 5M)
        self.net = EfficientNet.from_pretrained('efficientnet-b3')
        
        # Replace classifier
        n = self.net._fc.in_features
        self.net._fc = nn.Sequential(
            nn.Dropout(dropout),
            nn.Linear(n, 512),
            nn.ReLU(),
            nn.Dropout(dropout * 0.5),
            nn.Linear(512, num_classes)
        )
    
    def forward(self, x):
        return self.net(x)

model = EfficientNetB3(num_classes=14, dropout=0.4).to(device)

# Count parameters
total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"\n✅ EfficientNet-B3 Model:")
print(f"   Total parameters:     {total_params:,}")
print(f"   Trainable parameters: {trainable_params:,}")
print(f"   Model size:           ~{total_params * 4 / 1e6:.1f} MB")
```

```python
# Calculate class weights
class_counts = np.zeros(14)
for _, row in train_df.iterrows():
    for i, l in enumerate(labels):
        if l in row['Finding Labels']: class_counts[i] += 1

class_weights = len(train_df) / (14 * (class_counts + 1))
class_weights = torch.FloatTensor(class_weights).to(device)

print("\n📊 Class Weights:")
for i, label in enumerate(labels):
    print(f"   {label:20s}: {class_weights[i]:.2f}")

# Focal Loss
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma
    
    def forward(self, inputs, targets):
        BCE = nn.functional.binary_cross_entropy_with_logits(inputs, targets, reduction='none')
        pt = torch.exp(-BCE)
        focal = (1 - pt) ** self.gamma * BCE
        
        if self.alpha is not None:
            alpha_t = self.alpha.unsqueeze(0).expand_as(targets)
            focal = alpha_t * focal
        
        return focal.mean()

loss_fn = FocalLoss(alpha=class_weights, gamma=2)

# Optimizer with lower LR for bigger model
opt = torch.optim.AdamW(model.parameters(), lr=5e-5, weight_decay=1e-4)  # Lower LR

# Cosine annealing scheduler
scheduler = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts(opt, T_0=5, T_mult=1, eta_min=1e-6)

scaler = GradScaler(device='cuda')

print("\n✅ Training setup complete")
```

```python
def train_epoch(model, loader, loss_fn, opt, scaler, epoch, log_file):
    model.train()
    total_loss = 0
    labs, preds = [], []
    
    start_time = time.time()
    pbar = tqdm(loader, desc=f"Epoch {epoch:2d} Train", ncols=110, mininterval=1.0)
    
    for batch_idx, batch in enumerate(pbar):
        if batch is None: continue
        x, y = batch
        x, y = x.to(device), y.to(device)
        
        opt.zero_grad(set_to_none=True)
        
        with autocast(device_type='cuda', dtype=torch.float16):
            out = model(x)
            loss = loss_fn(out, y)
        
        scaler.scale(loss).backward()
        scaler.step(opt)
        scaler.update()
        
        total_loss += loss.item()
        labs.append(y.detach().cpu().numpy())
        preds.append(torch.sigmoid(out.detach()).cpu().numpy())
        
        # Update progress
        if batch_idx % 10 == 0:
            avg_loss = total_loss / (batch_idx + 1)
            pbar.set_postfix({'loss': f'{avg_loss:.4f}', 'lr': f'{opt.param_groups[0]["lr"]:.2e}'})
        
        # Log every 100 batches
        if batch_idx % 100 == 0:
            with open(log_file, 'a') as f:
                f.write(f"[{time.time():.0f}] Epoch {epoch} | Batch {batch_idx}/{len(loader)} | Loss {loss.item():.4f}\n")
    
    elapsed = time.time() - start_time
    auc = roc_auc_score(np.vstack(labs), np.vstack(preds), average='macro')
    
    return total_loss / len(loader), auc, elapsed

def val_epoch(model, loader, loss_fn):
    model.eval()
    total_loss = 0
    labs, preds = [], []
    
    with torch.no_grad():
        for batch in tqdm(loader, desc="Validating", ncols=110):
            if batch is None: continue
            x, y = batch
            x, y = x.to(device), y.to(device)
            
            with autocast(device_type='cuda', dtype=torch.float16):
                out = model(x)
                loss = loss_fn(out, y)
            
            total_loss += loss.item()
            labs.append(y.cpu().numpy())
            preds.append(torch.sigmoid(out).cpu().numpy())
    
    auc = roc_auc_score(np.vstack(labs), np.vstack(preds), average='macro')
    per_class_auc = roc_auc_score(np.vstack(labs), np.vstack(preds), average=None)
    
    return total_loss / len(loader), auc, per_class_auc
```

```python
log_file = 'phase5_effnet_b3_training.log'
results_file = 'phase5_effnet_b3_results.json'

with open(log_file, 'w') as f:
    f.write(f"Phase 5 Training Started: {time.ctime()}\n")
    f.write(f"Model: EfficientNet-B3\n")
    f.write(f"Target: Beat 0.82 AUC\n\n")

best_auc = 0
patience = 4
patience_counter = 0
results = []

print("\n" + "="*70)
print("🚀 STARTING EFFICIENTNET-B3 TRAINING")
print("="*70)
print(f"Start time: {time.strftime('%I:%M %p')}")
print(f"Target AUC: 0.82+")
print(f"Logging to: {log_file}")
print("="*70)

for epoch in range(1, 16):  # Max 15 epochs
    print(f"\n{'='*70}")
    print(f"EPOCH {epoch}/15")
    print(f"{'='*70}")
    
    # Train
    tr_loss, tr_auc, tr_time = train_epoch(model, train_loader, loss_fn, opt, scaler, epoch, log_file)
    
    # Validate
    val_loss, val_auc, per_class_auc = val_epoch(model, val_loader, loss_fn)
    
    # Step scheduler
    scheduler.step()
    
    # Print summary
    print(f"\n📊 Results:")
    print(f"   Train: Loss={tr_loss:.4f} | AUC={tr_auc:.4f} | Time={tr_time/60:.1f}min")
    print(f"   Val:   Loss={val_loss:.4f} | AUC={val_auc:.4f}")
    print(f"   LR:    {opt.param_groups[0]['lr']:.2e}")
    
    # Save best
    if val_auc > best_auc:
        best_auc = val_auc
        torch.save(model.state_dict(), "best_efficientnet_b3.pth")
        patience_counter = 0
        
        print(f"\n✅ NEW BEST MODEL! AUC={best_auc:.4f}")
        print(f"\n   Per-Class AUC:")
        for i, label in enumerate(labels):
            print(f"      {label:20s}: {per_class_auc[i]:.4f}")
        
        # Check if we beat target
        if best_auc >= 0.82:
            print(f"\n🎉 TARGET ACHIEVED! AUC = {best_auc:.4f} >= 0.82")
    else:
        patience_counter += 1
        print(f"   No improvement ({patience_counter}/{patience})")
    
    # Save results
    results.append({
        'epoch': epoch,
        'train_loss': tr_loss,
        'train_auc': tr_auc,
        'val_loss': val_loss,
        'val_auc': val_auc,
        'per_class_auc': per_class_auc.tolist(),
        'lr': opt.param_groups[0]['lr']
    })
    
    with open(results_file, 'w') as f:
        json.dump(results, f, indent=2)
    
    # Log
    with open(log_file, 'a') as f:
        f.write(f"[{time.time():.0f}] Epoch {epoch} | Train AUC: {tr_auc:.4f} | Val AUC: {val_auc:.4f} | Best: {best_auc:.4f}\n")
    
    # Early stopping
    if patience_counter >= patience:
        print(f"\n⏹️ Early stopping triggered at epoch {epoch}")
        break
    
    # Memory cleanup
    torch.cuda.empty_cache()

print("\n" + "="*70)
print("✅ TRAINING COMPLETE")
print("="*70)
print(f"End time: {time.strftime('%I:%M %p')}")
print(f"Best Val AUC: {best_auc:.4f}")
print(f"Model saved: best_efficientnet_b3.pth")
print(f"Results saved: {results_file}")
print("="*70)
```

```python
print("\n" + "="*70)
print("🧪 EVALUATING ON TEST SET")
print("="*70)

# Load best model
model.load_state_dict(torch.load("best_efficientnet_b3.pth"))
model.eval()

test_labs, test_preds = [], []

with torch.no_grad():
    for batch in tqdm(test_loader, desc="Testing"):
        if batch is None: continue
        x, y = batch
        x = x.to(device)
        
        with autocast(device_type='cuda', dtype=torch.float16):
            out = torch.sigmoid(model(x))
        
        test_labs.append(y.numpy())
        test_preds.append(out.cpu().numpy())

test_labs = np.vstack(test_labs)
test_preds = np.vstack(test_preds)

# Metrics
test_auc_macro = roc_auc_score(test_labs, test_preds, average='macro')
test_auc_weighted = roc_auc_score(test_labs, test_preds, average='weighted')
test_per_class = roc_auc_score(test_labs, test_preds, average=None)

print("\n" + "="*70)
print("TEST RESULTS: EfficientNet-B3")
print("="*70)
print(f"AUC (macro):    {test_auc_macro:.4f}")
print(f"AUC (weighted): {test_auc_weighted:.4f}")

print("\nPer-Class AUC:")
for i, label in enumerate(labels):
    print(f"  {label:20s}: {test_per_class[i]:.4f}")

# Comparison
print("\n" + "="*70)
print("PERFORMANCE COMPARISON")
print("="*70)
print(f"Baseline Ensemble (B0):        0.7870")
print(f"Balanced B0 (Phase 4):         0.7945")
print(f"EfficientNet-B3 (Phase 5):     {test_auc_macro:.4f}")
print(f"Improvement vs Balanced B0:    {(test_auc_macro - 0.7945)*100:+.2f}%")
print(f"Improvement vs Baseline:       {(test_auc_macro - 0.7870)*100:+.2f}%")

# Save
phase5_results = {
    'model': 'EfficientNet-B3',
    'test_auc_macro': test_auc_macro,
    'test_auc_weighted': test_auc_weighted,
    'per_class_auc': test_per_class.tolist(),
    'best_val_auc': best_auc
}

with open('phase5_test_results.json', 'w') as f:
    json.dump(phase5_results, f, indent=2)

print("\n✅ Test results saved to phase5_test_results.json")
print("="*70)
```

```python
import torch.nn.functional as F

print("\n" + "="*70)
print("APPLYING TEST-TIME AUGMENTATION (TTA)")
print("="*70)

def predict_with_tta(model, image_tensor, device, n_aug=4):
    """
    Predict with TTA:
    1. Original
    2. Horizontal flip
    3. Vertical shift
    4. Brightness adjustment
    """
    model.eval()
    predictions = []
    
    with torch.no_grad():
        # Original
        with torch.autocast(device_type='cuda', dtype=torch.float16):
            pred = torch.sigmoid(model(image_tensor.to(device)))
        predictions.append(pred.cpu())
        
        # Horizontal flip
        img_flipped = torch.flip(image_tensor, dims=[3])
        with torch.autocast(device_type='cuda', dtype=torch.float16):
            pred = torch.sigmoid(model(img_flipped.to(device)))
        predictions.append(pred.cpu())
        
        # Slight zoom (crop + resize)
        img_crop = F.interpolate(
            image_tensor[:, :, 10:214, 10:214], 
            size=(224, 224), 
            mode='bilinear'
        )
        with torch.autocast(device_type='cuda', dtype=torch.float16):
            pred = torch.sigmoid(model(img_crop.to(device)))
        predictions.append(pred.cpu())
        
        # Brightness adjustment (simulate lighting variation)
        img_bright = torch.clamp(image_tensor * 1.1, 0, 1)
        with torch.autocast(device_type='cuda', dtype=torch.float16):
            pred = torch.sigmoid(model(img_bright.to(device)))
        predictions.append(pred.cpu())
    
    # Average all predictions
    return torch.stack(predictions).mean(dim=0)

# Apply TTA to both models
print("Applying TTA to B0 model...")
b0_tta_preds = []
for batch in tqdm(test_loader, desc="B0 TTA"):
    if batch is None: continue
    x, _ = batch
    batch_preds = predict_with_tta(b0_model, x, device)
    b0_tta_preds.append(batch_preds.numpy())

b0_tta_preds = np.vstack(b0_tta_preds)

print("Applying TTA to B3 model...")
# Reload B3 model
b3_model = EfficientNetB3().to(device)
b3_model.load_state_dict(torch.load('best_efficientnet_b3.pth'))
b3_model.eval()

b3_tta_preds = []
for batch in tqdm(test_loader, desc="B3 TTA"):
    if batch is None: continue
    x, _ = batch
    batch_preds = predict_with_tta(b3_model, x, device)
    b3_tta_preds.append(batch_preds.numpy())

b3_tta_preds = np.vstack(b3_tta_preds)

# Ensemble with TTA
ensemble_tta_preds = 0.6 * b0_tta_preds + 0.4 * b3_tta_preds

# Evaluate
tta_auc = roc_auc_score(test_labs, ensemble_tta_preds, average='macro')
tta_per_class = roc_auc_score(test_labs, ensemble_tta_preds, average=None)

print("\n" + "="*70)
print("TTA RESULTS")
print("="*70)
print(f"Ensemble (no TTA):    0.8076")
print(f"Ensemble (with TTA):  {tta_auc:.4f}")
print(f"TTA Improvement:      {(tta_auc - 0.8076)*100:+.2f}%")

print("\nPer-Class AUC with TTA:")
for i, label in enumerate(labels):
    improvement = tta_per_class[i] - ensemble_per_class[i]
    print(f"  {label:20s}: {tta_per_class[i]:.4f} ({improvement:+.4f})")

# Save
tta_results = {
    'model': 'Ensemble B0+B3 with TTA',
    'test_auc': tta_auc,
    'per_class_auc': tta_per_class.tolist(),
    'tta_augmentations': ['original', 'h_flip', 'crop_zoom', 'brightness']
}

with open('ensemble_tta_results.json', 'w') as f:
    json.dump(tta_results, f, indent=2)

print("\n✅ TTA results saved to ensemble_tta_results.json")
print("="*70)
```
