```python
class ViTDataset(torch.utils.data.Dataset):
    def __init__(self, df, img_dir, transform):
        self.df = df.reset_index(drop=True)
        self.img_dir = img_dir
        self.transform = transform
        self.labels = ['Atelectasis','Consolidation','Infiltration','Pneumothorax','Edema',
                       'Emphysema','Fibrosis','Effusion','Pneumonia','Pleural_Thickening',
                       'Cardiomegaly','Nodule','Mass','Hernia']
    
    def __len__(self): return len(self.df)
    
    def __getitem__(self, i):
        try:
            row = self.df.iloc[i]
            img_path = None
            for j in range(1,13):
                p = os.path.join(self.img_dir, f'images_{j:03d}/images', row['Image Index'])
                if os.path.exists(p): img_path = p; break
            
            img = np.array(Image.open(img_path).convert('RGB'))
            img = self.transform(image=img)['image']
            
            y = np.zeros(len(self.labels), dtype=np.float32)
            if row['Finding Labels'] != "No Finding":
                for k, l in enumerate(self.labels):
                    if l in row['Finding Labels']: y[k] = 1
            
            return img, torch.tensor(y)
        except Exception:
            return None

def safe_collate(batch):
    batch = [b for b in batch if b is not None]
    if not batch: return None
    return torch.utils.data.dataloader.default_collate(batch)
```

```python
extract = "./chest_xray_data/"
data = pd.read_csv(os.path.join(extract, "Data_Entry_2017.csv"))

# Same split as before for fair comparison
train_df, val_df, test_df = np.split(
    data.sample(frac=1, random_state=42),
    [int(.7*len(data)), int(.85*len(data))]
)

# ViT works well with standard ImageNet preprocessing
tf_train = Compose([
    Resize(224, 224),
    HorizontalFlip(p=0.5),
    RandomBrightnessContrast(0.15, 0.15, p=0.3),
    Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225]),
    ToTensorV2()
])

tf_val = Compose([
    Resize(224, 224),
    Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225]),
    ToTensorV2()
])

train_ds = ViTDataset(train_df, extract, tf_train)
val_ds = ViTDataset(val_df, extract, tf_val)
test_ds = ViTDataset(test_df, extract, tf_val)

# Balanced sampler
labels = ['Atelectasis','Consolidation','Infiltration','Pneumothorax','Edema',
          'Emphysema','Fibrosis','Effusion','Pneumonia','Pleural_Thickening',
          'Cardiomegaly','Nodule','Mass','Hernia']

def create_sampler(df, labels):
    class_counts = np.zeros(len(labels))
    for i, label in enumerate(labels):
        class_counts[i] = df['Finding Labels'].str.contains(label, na=False).sum()
    
    class_weights = 1.0 / (class_counts + 1)
    class_weights = class_weights / class_weights.sum() * len(labels)
    
    sample_weights = []
    for idx, row in df.iterrows():
        labels_present = [i for i, l in enumerate(labels) if l in row['Finding Labels']]
        if labels_present:
            sample_weights.append(max(class_weights[labels_present]))
        else:
            sample_weights.append(1.0)
    
    return torch.utils.data.WeightedRandomSampler(sample_weights, len(sample_weights), replacement=True)

sampler = create_sampler(train_df, labels)

# DataLoaders - batch size 10 for ViT (memory intensive)
train_loader = DataLoader(train_ds, batch_size=10, sampler=sampler,
                          num_workers=0, collate_fn=safe_collate, pin_memory=False)
val_loader = DataLoader(val_ds, batch_size=10, shuffle=False,
                        num_workers=0, collate_fn=safe_collate, pin_memory=False)
test_loader = DataLoader(test_ds, batch_size=10, shuffle=False,
                         num_workers=0, collate_fn=safe_collate, pin_memory=False)

print(f"✅ Data loaded:")
print(f"   Train: {len(train_df):,} images ({len(train_loader)} batches)")
print(f"   Val:   {len(val_df):,} images ({len(val_loader)} batches)")
print(f"   Test:  {len(test_df):,} images ({len(test_loader)} batches)")
```

```python
class VisionTransformer(nn.Module):
    def __init__(self, num_classes=14, pretrained=True):
        super().__init__()
        
        # Load pretrained ViT-Base/16 from timm
        # vit_base_patch16_224: 86M parameters
        self.vit = timm.create_model(
            'vit_base_patch16_224',
            pretrained=pretrained,
            num_classes=0  # Remove classification head
        )
        
        # Get embedding dimension
        self.embed_dim = self.vit.num_features  # 768 for ViT-Base
        
        # Custom classification head for multi-label
        self.classifier = nn.Sequential(
            nn.LayerNorm(self.embed_dim),
            nn.Dropout(0.3),
            nn.Linear(self.embed_dim, 512),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(512, num_classes)
        )
    
    def forward(self, x):
        # Extract features
        features = self.vit(x)  # [batch, 768]
        # Classify
        return self.classifier(features)

print("Loading ViT model...")
model = VisionTransformer(num_classes=14, pretrained=True).to(device)

# Count parameters
total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"\n✅ Vision Transformer Model:")
print(f"   Architecture: ViT-Base/16")
print(f"   Image size: 224×224")
print(f"   Patch size: 16×16")
print(f"   Total parameters: {total_params:,}")
print(f"   Trainable parameters: {trainable_params:,}")
print(f"   Model size: ~{total_params * 4 / 1e6:.1f} MB")
```

```python
# Class weights
class_counts = np.zeros(14)
for _, row in train_df.iterrows():
    for i, l in enumerate(labels):
        if l in row['Finding Labels']: class_counts[i] += 1

class_weights = len(train_df) / (14 * (class_counts + 1))
class_weights = torch.FloatTensor(class_weights).to(device)

print("\n📊 Class Weights:")
for i, label in enumerate(labels):
    print(f"   {label:20s}: {class_weights[i]:.2f}")

# Focal Loss
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma
    
    def forward(self, inputs, targets):
        BCE = nn.functional.binary_cross_entropy_with_logits(inputs, targets, reduction='none')
        pt = torch.exp(-BCE)
        focal = (1 - pt) ** self.gamma * BCE
        
        if self.alpha is not None:
            alpha_t = self.alpha.unsqueeze(0).expand_as(targets)
            focal = alpha_t * focal
        
        return focal.mean()

loss_fn = FocalLoss(alpha=class_weights, gamma=2)

# Optimizer - ViT typically needs lower LR than CNNs
opt = torch.optim.AdamW(
    [
        {'params': model.vit.parameters(), 'lr': 1e-5},  # Pretrained ViT: very low LR
        {'params': model.classifier.parameters(), 'lr': 1e-4}  # New head: higher LR
    ],
    weight_decay=1e-4
)

# Cosine annealing with warmup
from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts

scheduler = CosineAnnealingWarmRestarts(opt, T_0=5, T_mult=1, eta_min=1e-7)

scaler = GradScaler(device='cuda')

print("\n✅ Training setup complete")
print(f"   Base LR (ViT): 1e-5")
print(f"   Head LR: 1e-4")
print(f"   Scheduler: CosineAnnealingWarmRestarts")
```

```python
def train_epoch(model, loader, loss_fn, opt, scaler, scheduler, epoch, log_file):
    model.train()
    total_loss = 0
    labs, preds = [], []
    
    start_time = time.time()
    pbar = tqdm(loader, desc=f"Epoch {epoch:2d} Train", ncols=110, mininterval=1.0)
    
    for batch_idx, batch in enumerate(pbar):
        if batch is None: continue
        x, y = batch
        x, y = x.to(device), y.to(device)
        
        opt.zero_grad(set_to_none=True)
        
        with autocast(device_type='cuda', dtype=torch.float16):
            out = model(x)
            loss = loss_fn(out, y)
        
        scaler.scale(loss).backward()
        
        # Gradient clipping for stability
        scaler.unscale_(opt)
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        
        scaler.step(opt)
        scaler.update()
        
        total_loss += loss.item()
        labs.append(y.detach().cpu().numpy())
        preds.append(torch.sigmoid(out.detach()).cpu().numpy())
        
        # Update progress
        if batch_idx % 10 == 0:
            avg_loss = total_loss / (batch_idx + 1)
            lr_vit = opt.param_groups[0]['lr']
            lr_head = opt.param_groups[1]['lr']
            pbar.set_postfix({
                'loss': f'{avg_loss:.4f}', 
                'lr_vit': f'{lr_vit:.2e}',
                'lr_head': f'{lr_head:.2e}'
            })
        
        # Log every 100 batches
        if batch_idx % 100 == 0:
            with open(log_file, 'a') as f:
                f.write(f"[{time.time():.0f}] Epoch {epoch} | Batch {batch_idx}/{len(loader)} | Loss {loss.item():.4f}\n")
    
    elapsed = time.time() - start_time
    auc = roc_auc_score(np.vstack(labs), np.vstack(preds), average='macro')
    
    return total_loss / len(loader), auc, elapsed

def val_epoch(model, loader, loss_fn):
    model.eval()
    total_loss = 0
    labs, preds = [], []
    
    with torch.no_grad():
        for batch in tqdm(loader, desc="Validating", ncols=110):
            if batch is None: continue
            x, y = batch
            x, y = x.to(device), y.to(device)
            
            with autocast(device_type='cuda', dtype=torch.float16):
                out = model(x)
                loss = loss_fn(out, y)
            
            total_loss += loss.item()
            labs.append(y.cpu().numpy())
            preds.append(torch.sigmoid(out).cpu().numpy())
    
    auc = roc_auc_score(np.vstack(labs), np.vstack(preds), average='macro')
    per_class_auc = roc_auc_score(np.vstack(labs), np.vstack(preds), average=None)
    
    return total_loss / len(loader), auc, per_class_auc
```

```python
log_file = 'phase6_vit_training.log'
results_file = 'phase6_vit_results.json'

with open(log_file, 'w') as f:
    f.write(f"Phase 6 ViT Training Started: {time.ctime()}\n")
    f.write(f"Model: Vision Transformer Base/16\n")
    f.write(f"Target: Ensemble AUC 0.82-0.83\n\n")

best_auc = 0
patience = 4
patience_counter = 0
results = []

print("\n" + "="*70)
print("🚀 STARTING VISION TRANSFORMER TRAINING")
print("="*70)
print(f"Start time: {time.strftime('%I:%M %p')}")
print(f"Target: Contribute to 0.82+ AUC ensemble")
print(f"Logging to: {log_file}")
print("="*70)

for epoch in range(1, 16):  # Max 15 epochs
    print(f"\n{'='*70}")
    print(f"EPOCH {epoch}/15")
    print(f"{'='*70}")
    
    # Train
    tr_loss, tr_auc, tr_time = train_epoch(
        model, train_loader, loss_fn, opt, scaler, scheduler, epoch, log_file
    )
    
    # Validate
    val_loss, val_auc, per_class_auc = val_epoch(model, val_loader, loss_fn)
    
    # Step scheduler
    scheduler.step()
    
    # Print summary
    print(f"\n📊 Results:")
    print(f"   Train: Loss={tr_loss:.4f} | AUC={tr_auc:.4f} | Time={tr_time/60:.1f}min")
    print(f"   Val:   Loss={val_loss:.4f} | AUC={val_auc:.4f}")
    print(f"   LR (ViT):  {opt.param_groups[0]['lr']:.2e}")
    print(f"   LR (Head): {opt.param_groups[1]['lr']:.2e}")
    
    # Save best
    if val_auc > best_auc:
        best_auc = val_auc
        torch.save(model.state_dict(), "best_vit_base.pth")
        patience_counter = 0
        
        print(f"\n✅ NEW BEST MODEL! AUC={best_auc:.4f}")
        print(f"\n   Per-Class AUC:")
        for i, label in enumerate(labels):
            print(f"      {label:20s}: {per_class_auc[i]:.4f}")
    else:
        patience_counter += 1
        print(f"   No improvement ({patience_counter}/{patience})")
    
    # Save results
    results.append({
        'epoch': epoch,
        'train_loss': tr_loss,
        'train_auc': tr_auc,
        'val_loss': val_loss,
        'val_auc': val_auc,
        'per_class_auc': per_class_auc.tolist(),
        'lr_vit': opt.param_groups[0]['lr'],
        'lr_head': opt.param_groups[1]['lr']
    })
    
    with open(results_file, 'w') as f:
        json.dump(results, f, indent=2)
    
    # Log
    with open(log_file, 'a') as f:
        f.write(f"[{time.time():.0f}] Epoch {epoch} | Train AUC: {tr_auc:.4f} | Val AUC: {val_auc:.4f} | Best: {best_auc:.4f}\n")
    
    # Early stopping
    if patience_counter >= patience:
        print(f"\n⏹️ Early stopping triggered at epoch {epoch}")
        break
    
    # Memory cleanup
    torch.cuda.empty_cache()

print("\n" + "="*70)
print("✅ ViT TRAINING COMPLETE")
print("="*70)
print(f"End time: {time.strftime('%I:%M %p')}")
print(f"Best Val AUC: {best_auc:.4f}")
print(f"Model saved: best_vit_base.pth")
print(f"Results saved: {results_file}")
print("="*70)
```

```python
# ViT test evaluation (fixed)
import numpy as np
from sklearn.metrics import roc_auc_score

print("\n" + "="*70)
print("🧪 EVALUATING ViT ON TEST SET (FIXED)")
print("="*70)

model.load_state_dict(torch.load("best_vit_base.pth"))
model.eval()

vit_test_preds_list = []
vit_test_labs_list  = []   # <-- collect labels

with torch.no_grad():
    for batch in tqdm(test_loader, desc="Testing ViT"):
        if batch is None:
            continue
        x, y = batch
        x = x.to(device)

        with autocast(device_type='cuda', dtype=torch.float16):
            out = torch.sigmoid(model(x))

        vit_test_preds_list.append(out.cpu().numpy())
        vit_test_labs_list.append(y.numpy())   # <-- append labels

# Stack to arrays of shape (n_samples, 14)
vit_test_preds = np.vstack(vit_test_preds_list).astype(np.float32)  # scores y_score
test_labs      = np.vstack(vit_test_labs_list).astype(np.int32)     # labels y_true

# Metrics
vit_auc_macro    = roc_auc_score(test_labs, vit_test_preds, average='macro')
vit_auc_weighted = roc_auc_score(test_labs, vit_test_preds, average='weighted')
vit_per_class    = roc_auc_score(test_labs, vit_test_preds, average=None)

print("\n" + "="*70)
print("TEST RESULTS: Vision Transformer")
print("="*70)
print(f"AUC (macro):    {vit_auc_macro:.4f}")
print(f"AUC (weighted): {vit_auc_weighted:.4f}")
print("\nPer-Class AUC:")
for i, label in enumerate(labels):
    print(f"  {label:20s}: {vit_per_class[i]:.4f}")
```

```python
# Rebuild EfficientNet-B3 predictions as `test_preds`
import numpy as np, torch, torch.nn as nn
from tqdm.auto import tqdm

class EfficientNetB3(nn.Module):
    def __init__(self, num_classes=14, dropout=0.4):
        super().__init__()
        from efficientnet_pytorch import EfficientNet
        self.net = EfficientNet.from_pretrained('efficientnet-b3')
        n = self.net._fc.in_features
        self.net._fc = nn.Sequential(
            nn.Dropout(dropout),
            nn.Linear(n, 512),
            nn.ReLU(),
            nn.Dropout(dropout * 0.5),
            nn.Linear(512, num_classes)
        )
    def forward(self, x): return self.net(x)

b3_model = EfficientNetB3().to(device)
b3_model.load_state_dict(torch.load("best_efficientnet_b3.pth"))
b3_model.eval()

b3_pred_list = []
with torch.no_grad():
    for batch in tqdm(test_loader, desc="B3 Predictions"):
        if batch is None: continue
        x, _ = batch
        with torch.autocast(device_type='cuda', dtype=torch.float16):
            out = torch.sigmoid(b3_model(x.to(device)))
        b3_pred_list.append(out.cpu().numpy())

test_preds = np.vstack(b3_pred_list).astype(np.float32)  # B3 probabilities
print("B3 test_preds shape:", test_preds.shape)  # (n_samples, 14)
```

```python
from sklearn.metrics import roc_auc_score

weight_configs = [
    (0.40, 0.30, 0.30),  # B0, B3, ViT
    (0.50, 0.25, 0.25),
    (0.40, 0.20, 0.40),
    (0.33, 0.33, 0.34),
]

best_auc, best_w, best_preds = -1.0, None, None
print("\nTesting 3-way weights...")
for w_b0, w_b3, w_vit in weight_configs:
    ens = w_b0 * b0_preds + w_b3 * test_preds + w_vit * vit_test_preds
    auc_macro = roc_auc_score(test_labs, ens, average='macro')
    print(f"  Weights (B0={w_b0:.2f}, B3={w_b3:.2f}, ViT={w_vit:.2f}) → AUC={auc_macro:.4f}")
    if auc_macro > best_auc:
        best_auc, best_w, best_preds = auc_macro, (w_b0, w_b3, w_vit), ens

print(f"\n✅ Best 3-way ensemble AUC: {best_auc:.4f} with weights B0={best_w[0]:.2f}, B3={best_w[1]:.2f}, ViT={best_w[2]:.2f}")
```

```python

# Get per-class AUC for the best ensemble
best_per_class = roc_auc_score(test_labs, best_preds, average=None)

print("\n" + "="*70)
print("FINAL 3-WAY ENSEMBLE: PER-CLASS RESULTS")
print("="*70)

comparison_table = []
for i, label in enumerate(labels):
    comparison_table.append({
        'Disease': label,
        'B0': b0_preds[:, i].mean(),  # avg prediction confidence
        'B3': test_preds[:, i].mean(),
        'ViT': vit_test_preds[:, i].mean(),
        'Ensemble AUC': best_per_class[i]
    })

import pandas as pd
comp_df = pd.DataFrame(comparison_table)
comp_df = comp_df.sort_values('Ensemble AUC', ascending=False)

print(comp_df.to_string(index=False))

# Minority class summary
minority_classes = ['Hernia', 'Pneumonia', 'Edema', 'Emphysema', 'Fibrosis']
minority_aucs = [best_per_class[labels.index(c)] for c in minority_classes]

print("\n" + "="*70)
print("MINORITY CLASS PERFORMANCE")
print("="*70)
for c in minority_classes:
    idx = labels.index(c)
    print(f"{c:20s}: {best_per_class[idx]:.4f}")
print(f"\nAverage Minority AUC: {np.mean(minority_aucs):.4f}")

# Save final results
final_results = {
    'model': '3-Way Ensemble (B0 + B3 + ViT)',
    'weights': {'B0': 0.40, 'B3': 0.20, 'ViT': 0.40},
    'test_auc_macro': best_auc,
    'per_class_auc': best_per_class.tolist(),
    'component_performance': {
        'B0_alone': 0.7945,
        'B3_alone': 0.7884,
        'ViT_alone': 0.7931,
        'B0_B3_ensemble': 0.8076
    },
    'improvement_vs_baseline': (best_auc - 0.7870) * 100,
    'minority_class_avg': np.mean(minority_aucs)
}

import json
with open('final_3way_ensemble_results.json', 'w') as f:
    json.dump(final_results, f, indent=2)

print("\n✅ Final results saved to final_3way_ensemble_results.json")
```
